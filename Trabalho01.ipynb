**Dados obtidos**

Ao analisar os dados obtidos, notou-se que a idade seria uma informação relevante a ser exibida, com o objetivo de identificar a faixa etária em que a programação é mais popular. *Ambito Trabalhista*

Informações relacionadas ao contexto de trabalho dos participantes que responderam a pesquisa também foram disponibilizados, de maneira que foi possível identificar algumas coisas. Foram plotados dois gráficos de pizza (ou torta, do inglês) para representar essas informações. Um deles refere-se a modalidade de trabalho dos individuos, sendo a Hibrida e Remota a grande maiora. Isso reflete a crescente tendência de trabalhos realizados a partir de casa, impulsionada pelo desenvolvimento de programação e das ferramentas que possibilitam o trabalho remoto ou híbrido. Além disso, no que diz respeito a empregabilidade, nota-se que 73,4% dos participantes da pesquisa estavam empregados no momento da realização da pesquisa. 

```python
def situacao(sit):
    if pd.isna(sit):
        return "No information"
    elif "Employed, full-time" in sit or "Employed, part-time" in sit:
        return "Employed"
    elif "I prefer not to say" in sit:
        return "Without comments"
    else:
        return "unemployed"

RemoteWork = df['RemoteWork'].value_counts()
dE = df
dE['Situation'] = dE['Employment'].apply(situacao)
dE['Situation'] 
Employee = dE[dE["Situation"] == 'Employed']
Unemployed = dE[dE["Situation"] == 'unemployed']
Without_information = dE[(dE["Situation"] == 'Without comments') | (dE["Situation"] == 'No information')]
Employee = len(Employee)
Unemployed = len(Unemployed)
Without_information = len(Without_information)
```

```python
RemoteWork = df['RemoteWork'].value_counts()
plt.rcParams['figure.figsize'] = [10,6]
sns.set_style('darkgrid')

colors_modalidade = ['#008080', '#00CED1', '#3CB371']
colors_emprego = ['#008080', '#00CED1', '#3CB371']

plt.subplot(1,2,1)
plt.pie([RemoteWork['Remote'],RemoteWork['Hybrid (some remote, some in-person)'],RemoteWork['In-person']],labels=['Remota','Híbrida','Presencial'],shadow=True,autopct='%1.1f%%',startangle=90, colors=colors_modalidade)
plt.title("Modalidade de Trabalho dos Participantes")
plt.subplot(1,2,2)
plt.pie([Employee,Unemployed,Without_information],labels=['Empregado','Desempregado','Sem Informação'],shadow=True,autopct='%1.1f%%',startangle=90, colors=colors_emprego)
plt.title("Status de Emprego dos Participantes")
plt.show()
```

*Ensino Superior*

Sabe-se que os profissionais de Tecnologia da Informação (TI) não têm um salário mínimo fixo, o que pode influenciar a decisão de indivíduos que trabalham na área em relação à busca ou não de ensino superior. A partir dos dados coletados em uma das perguntas da pesquisa, observou-se que, embora o ensino superior não seja um requisito absoluto para atuar na área, a maioria dos profissionais de programação possui formação universitária. Para facilitar a visualização desses dados, foi criado um gráfico de pizza.

```python
df2 = df[df['MainBranch'] == 'I am a developer by profession']

included_levels = ["Bachelor’s degree (B.A., B.S., B.Eng., etc.)", "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)","Some college/university study without earning a degree", "Professional degree (JD, MD, Ph.D, Ed.D, etc.)", "Associate degree (A.A., A.S., etc.)"]

included_data = df2[df2['EdLevel'].isin(included_levels)]

excluded_data = df2[~df2['EdLevel'].isin(included_levels)]

Superior = included_data['EdLevel'].count()
notSuperior = excluded_data['EdLevel'].count()

notSuperior

colors_ensup = ['#008080', '#00CED1']

plt.rcParams['figure.figsize'] = [10,6]
sns.set_style('darkgrid')
plt.pie([Superior,notSuperior],labels=['Possuem algum Ensino Superior','Não Possuem Ensino Superior'],shadow=True,autopct='%1.1f%%',startangle=90, colors=colors_ensup)
plt.title("Trabalhadores com Formação em Ensino Superior")
plt.show()
```

*Linguagens de Programação*

Um ponto fundamental a ser destacado é a variedade de linguagens de programação disponíveis para uso em computadores. Durante o decorrer da pesquisa do site Stack Overflow, essa questão foi abordada com o propósito de coletar dados sobre as tecnologias utilizadas pelos participantes. Como resultado, os autores elaboraram uma tabela (os valores são a quantidade de indivíduos em números), a partir dos dados disponibilizados no DataFrame obtido do arquivo csv, que oferece uma visão mais clara das porcentagens de indivíduos, divididos por faixa etária, que empregam as diferentes tecnologias mencionadas pelos entrevistados.

```python
df4 = df
df4['LanguageHaveWorkedWith'] = df4['LanguageHaveWorkedWith'].str.split(";")
df4 = df4.explode('LanguageHaveWorkedWith')
```

```python
language_usage = df4.groupby(['Age', 'LanguageHaveWorkedWith'])['LanguageHaveWorkedWith'].count().unstack(fill_value=0)
language_usage = language_usage.reindex(language_usage.sum().sort_values(ascending=False).index, axis=1)
language_usage
top_languages = language_usage.loc[:, language_usage.sum().sort_values(ascending=False).index[:]]
top_languages 
```

Nesta seção, são criados gráficos que representam o uso de diferentes linguagens de programação, permitindo a visualização da distribuição por faixa etária. As linguagens a serem exibidas nos gráficos podem ser personalizadas de acordo com as preferências de quem as visualiza. Basta ajustar a atribuição das variáveis l1, l2, l3 e l4 no campo de código abaixo. 

```python
top_languages_por = top_languages.div(top_languages.sum(axis=1), axis=0) * 100
plt.rcParams["figure.figsize"] = [12,8]
sns.set_context("notebook", font_scale=1, rc={"font.size":14,"axes.titlesize":16,"axes.labelsize":12})

fig, axes = plt.subplots(2, 2)

l1 = "Python"
l2 = "C#"
l3 = "C++"
l4 = "JavaScript"
y_label = "Quantidade de pessoas (%)"

top_languages_por[0:6][l1].plot(kind="bar",stacked=False, ax = axes[0,0], color='#008080')
axes[0, 0].set_title(l1)
axes[0, 0].set_ylabel(y_label)

top_languages_por[0:6][l2].plot(kind="bar",stacked=False, ax = axes[0,1], color='#008080')
axes[0, 1].set_title(l2)
axes[0, 1].set_ylabel(y_label)

top_languages_por[0:6][l3].plot(kind="bar",stacked=False, ax = axes[1,0], color='#008080')
axes[1, 0].set_title(l3)
axes[1, 0].set_ylabel(y_label)

top_languages_por[0:6][l4].plot(kind="bar",stacked=False, ax = axes[1,1], color='#008080')
axes[1, 1].set_title(l4)
axes[1, 1].set_ylabel(y_label)

plt.tight_layout()
plt.show()
```
